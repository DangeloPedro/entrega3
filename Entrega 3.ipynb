```python
driver.get('https://www.leagueofgraphs.com/pt/tft/summoners')
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
table = soup.table
table_body = table.find('tbody')

rows = table_body.find_all('tr')
data = []

for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip().replace('\n', '') for ele in cols]
    data.append([ele for ele in cols if ele])
```

```python
data
df = pd.DataFrame(data)
df = df.set_axis(['Rank', 'Nome de usuário', 'Divisão', 'Vitórias', 'Torneios_Jogados'], axis =1 )
df = df.drop(0)
```

```python
df['Pontos da liga'] = df.Divisão.str.extract('(\d+)')
df['Vitórias'] = df.Vitórias.str.extract('(\d+)')
df['Torneios_Jogados'] = df.Torneios_Jogados.str.extract('(\d+)')
```
